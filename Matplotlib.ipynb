```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,11)
print(x)
y=2*x
print(y)
plt.plot(x,y,color='yellow',linestyle=':',linewidth=5)
plt.title("line plot")
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,20)
print(x)
y1=5*x
print(y1)
y2=7*x
print(y2)
plt.plot(x,y1,color='g')
plt.plot(x,y2,color='y')
plt.title("line plot")
plt.xlabel("x_label")
plt.ylabel("y_plot")
plt.grid()
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,20)
print(x)
y1=5*x
print(y1)
y2=7*x
print(y2)
plt.subplot(1,2,1)
plt.plot(x,y1,color='g')
plt.title("line plot")
plt.xlabel("x_label")
plt.ylabel("y_plot")
plt.grid()
plt.subplot(1,2,2)
plt.plot(x,y2,color='y')
plt.title("line plot")
plt.xlabel("x_label")
plt.ylabel("y_plot")
plt.grid()
plt.show
```

```python
import numpy as np
student={"vanashree":8.74,"vijeta":6.94,"netra":8.54}
name=list(student.keys())
cgpa=list(student.values())
plt.bar(name,cgpa,color=['yellow','y','b'])
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.title("line_plot")
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[10,20,30,40,50,60]
y=[6,3,8,2,9,2]
plt.scatter(x,y,marker='*',c='g')
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[10,20,30,40,50,60]
y=[6,3,8,2,9,2]
y1=[3,7,2,9,6,3]
plt.scatter(x,y,marker='*',c='g',s=100)
plt.scatter(x,y1,marker='.',c='b',s=100)

plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[10,20,30,40,50,60]
y=[6,3,8,2,9,2]
y1=[3,7,2,9,6,3]
plt.subplot(1,2,1)
plt.scatter(x,y,marker='*',c='g',s=100)
plt.subplot(1,2,2)
plt.scatter(x,y1,marker='.',c='b',s=100)
plt.show
```
